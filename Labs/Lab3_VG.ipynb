```python
import pandas as pd
import matplotlib.pyplot as plt

def classify_points(k, m, df_values):
    global label_list
    label_list = []

    for i, j in df_values:
        if j > (k*i + m):
            e = 1
            label_list.append(e)
        else:
            e = 0
            label_list.append(e)
    return label_list

data = pd.read_csv("unlabelled_data.csv", header=None)
df = pd.DataFrame(data)
df_values = df.values
```

```python
# linear function f(x) = -0.489*x 
classify_points(-0.489,0, df_values) 
df['label'] = label_list

df_1 = df.loc[df["label"] == 0]
df_2 = df.loc[df["label"] == 1]
print(f"Total number of points classified as 0: {len(df_1)}")
print(f"Total number of points classified as 1: {len(df_2)}")
    

# plots 
a = data.iloc[:, 0]
b = data.iloc[:, 1]
f = lambda a: -0.489*a
a1, a2 = df_1.iloc[:, 0], df_2.iloc[:, 0]
b1, b2 = df_1.iloc[:, 1], df_2.iloc[:, 1]

fig, ax = plt.subplots()
ax.scatter(a1, b1, label='0') 
ax.scatter(a2, b2, label='1')
ax.plot(a,f(a),label='-0.489*x')
ax.set(title="Labelled data", xlabel='x', ylabel='y')
ax.legend(fontsize=10, loc='best')


# linear function g(x) = -2*x + 0.16
classify_points(-2, 0.16, df_values) 
df['label'] = label_list

df_1 = df.loc[df["label"] == 0]
df_2 = df.loc[df["label"] == 1]
print(f"Total number of points classified as 0: {len(df_1)}")
print(f"Total number of points classified as 1: {len(df_2)}")

# plots 
g = lambda a: -2*a + 0.16
a1, a2 = df_1.iloc[:, 0], df_2.iloc[:, 0]
b1, b2 = df_1.iloc[:, 1], df_2.iloc[:, 1]

fig, ax = plt.subplots()
ax.scatter(a1, b1, label='0') 
ax.scatter(a2, b2, label='1')
ax.plot(a,g(a),label='y=-2*x+0.16')
ax.set(title="Labelled data", xlabel='x', ylabel='y')
ax.legend(fontsize=10, loc='best')

# linear function h(x) = 800*x - 120
classify_points(800, -120, df_values) 
df['label'] = label_list

df_1 = df.loc[df["label"] == 0]
df_2 = df.loc[df["label"] == 1]
print(f"Total number of points classified as 0: {len(df_1)}")
print(f"Total number of points classified as 1: {len(df_2)}")

# plots 
h = lambda a: 800*a -120
a1, a2 = df_1.iloc[:, 0], df_2.iloc[:, 0]
b1, b2 = df_1.iloc[:, 1], df_2.iloc[:, 1]

fig, ax = plt.subplots()
ax.scatter(a1, b1, label='0') 
ax.scatter(a2, b2, label='1')
ax.plot(a,h(a),label='y=800*x-120')
ax.set(title="Labelled data", xlabel='x', ylabel='y')
ax.legend(fontsize=10, loc='best')
```
