```python
# B. Rätta linjens ekvation

data = pd.read_csv("unlabelled_data.csv", header=None)
a = data.iloc[:, 0]
b = data.iloc[:, 1]

x1, x2 = 4, -4
y1, y2 = -4, 4

k = (y2-y1)/(x2-x1)
m = y1 - k*x1
y = [k*x + m for x in a]

fig, ax = plt.subplots()
ax.plot(a,y,'-')
ax.scatter(a,b)
```

```python
import pandas as pd

data = pd.read_csv("unlabelled_data.csv",header=None)
df = data.values

def points(k,m,df):
    label_list = []

    for i, j in df:
        if j > (k*i + m):
            e = 1
            label_list.append(e)
        else: 
            e = 0
            label_list.append(e)
    return label_list

points(-1,0,df)
```

```python
# D. En funktion som räknar ut om en punkt ligger till höger om/ovanför eller vänster om/nedanför linjen 
import pandas as pd

data = pd.read_csv("unlabelled_data.csv",header=None)

def points(k,m,data):
    
    label_list = []

    for i, j in data.values:
        if j > (-k*i + m):
            e = 1
            label_list.append(e)
        else: 
            e = 0
            label_list.append(e)
    return label_list

#points(-2, 0.16, data)
points(-0.489,0,data)
```

```python

# D. En funktion som räknar ut om en punkt ligger till höger om/ovanför eller vänster om/nedanför linjen 
import pandas as pd

data = pd.read_csv("unlabelled_data.csv",header=None)

label_list = []

for i, j in data.values:
    if j > (-i):
        e = 1
        label_list.append(e)
    else: 
        e = 0
        label_list.append(e)

df = pd.DataFrame(data)
df['label'] = label_list
df.to_csv("../labelled_data.csv", index=False)

print(df)
```
