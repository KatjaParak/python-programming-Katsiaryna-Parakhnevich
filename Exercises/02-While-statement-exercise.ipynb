```python
#2a. Arithmetic sum

i = 0
sum = 0

while i < 100:
    i += 1
    sum += i

    print(f"{i}", end="")
    if i != 100:
        print("+", end="")
    if i == 100:
        print(f"={sum}", end="")
```

```python
#3a. Guess number game

import random as rnd 

random_number = rnd.randint(1, 100)
guesses = 0

while True:
    user_number = int(input('Enter a number between 1 and 100: '))
    guesses += 1
    if user_number == random_number:
        print(f'Congratulations! You got it! Correct number is {user_number}.')
        print(f"You've guessed from {guesses}th attempt. Game over.")
        break
    else:
        print(f'{user_number} is wrong. Try again')
    print(f'Total number of guesses is {guesses}.')
    if user_number > random_number:
        print('Your number is too high.')
    else:
        print('Your number is too low.')
```

```python
#3b. Guess number game 
import random as rnd

r_number = rnd.randint(1, 100)
attempts = 0
low = 1
high = 100

while True:
    c_number = rnd.randint(1, 100)
    attempts += 1
    if c_number == r_number:
        print(f'The computer guessed {guess} and it was correct!')
        print(f'Game over. Total number of guesses is {attempts}.')
        break
    else:
        guess = (low+high)//2
        if guess < c_number:
            low += guess 
        elif guess > c_number:
            high -= guess
        print(f'The computer takes a guess...{guess}')
```

```python
#4a. Multiplication game

score = 0

while True:
    x = rnd.randint(1, 10)
    y = rnd.randint(1, 10)

    user_answer = int(input(f'What is {x} x {y} ?'))
    if user_answer == x*y:
        print("Good work!")
        score += 1
    else:
        print(f"Answer is wrong. The correct answer is {x*y}")

    play_again = input("Do you want to play again (y for yes)?")
    if play_again != "y":
        print("Thanks for game! Game over.")
        break
```

```python
#5b Check convergence

import numpy as np

def term(n):
    return np.divide(np.float_power((-1), n), 2*n+1)

n = 0
acc = 0
v = 0
e = 1
while e > np.float_power(10, -6):
    prev = v
    v = term(n)
    e = np.absolute(prev - v)
    acc += v
    n += 1

print(
    f"Number of iterations: {n}\n sum: {acc:.4f}  or {np.divide(acc, np.pi):.4f}pi ")
```
